## Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python


def nhl_correlation():
    import pandas as pd
    import numpy as np
    import scipy.stats as stats
    import re

    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    
    ##Cleaning the cities df
    cities['NHL'].replace(r'(.*)\[.*\].*|[-—]',r'\1',regex=True, inplace=True)
    cities['NHL'].replace('', np.nan, inplace=True)
    cities.dropna(inplace=True)
    cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
    cities["NHL"] = cities["NHL"].replace({"RangersIslandersDevils": "Rangers Islanders Devils", "KingsDucks": "Kings Ducks"})
    cities.drop(['NFL','MLB','NBA'], axis=1,inplace=True)

    ##cleaning the nhl_df
    nhl = nhl_df[nhl_df['year'] == 2018]
    nhl = nhl[nhl["team"].str.contains('Atlantic Division|Metropolitan Division|Central Division|Pacific Division') == False]
    nhl.drop(['League','year','RPt%','ROW','PTS%','PTS','GP','OL','GF','GA','SRS','SOS'],axis=1, inplace=True)
    nhl['team'].replace(r'(.*)[\*].*', r"\1".strip(),regex=True, inplace=True)
    
    ##split the city from the team names prior to merging
    nhl['only teams'] = nhl['team'].apply(lambda x: x.rsplit(None, 1)[-1])
    nhl['only teams'] = nhl['only teams'].str.strip()
    nhl['only teams'] = nhl['only teams'].replace({'Leafs':'Maple Leafs','Wings':'Red Wings',
                                                  'Jackets':'Blue Jackets','Knights':'Golden Knights'})
    ##changes columns to numeric
    cols = ['W','L']
    nhl[cols] = nhl[cols].apply(pd.to_numeric, errors='coerce', axis=1)
    
    ##takes the cities with multiple teams from the nhl df so averages can be taken
    newYork = nhl[nhl['only teams'].isin(['Devils','Islanders','Rangers'])]
    LA = nhl[nhl['only teams'].isin(['Kings','Ducks'])]
    nhl = nhl[nhl['only teams'].str.contains('Devils|Islanders|Rangers|Kings|Ducks')==False]

    ##Find the average values for the New York teams
    newYork.loc['mean'] = newYork.mean()
    newYork = newYork.iloc[3:4]
    newYork['team'] = newYork['team'].fillna('New York City')
    newYork['only teams'] = newYork['only teams'].fillna('Rangers Islanders Devils')
    
    ##Find the average values for the LA teams
    LA.loc['mean'] = LA.mean()
    LA = LA.iloc[2:3]
    LA['team'] = LA['team'].fillna('Los Angeles')
    LA['only teams'] = LA['only teams'].fillna('Kings Ducks')

    ##Merge the nhl dfs back together
    nhl = pd.merge(nhl, newYork, how='outer')
    nhl = pd.merge(nhl, LA, how='outer')
    
    # merge the dataframes, drop the unnecessary columns, calculate the win-loss ratio
    df = pd.merge(cities, nhl, left_on="NHL", right_on="only teams",how='outer')
    df.drop(['only teams','team'], axis=1, inplace=True)
    df['W/L Ratio'] = df['W']/(df['W']+df['L'])
    df['Population'] = df['Population'].apply(pd.to_numeric, errors='coerce')
    df.set_index('Metropolitan area',inplace=True)

    
    population_by_region = df['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = df['W/L Ratio'] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```
